```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

columns = ['ID', 'Semaine', 'Province', 'Pluviometrie_mm', 'Temperature_C', 'NDVI',
       'Humidite_pct', 'Densite_Population', 'Couverture_Vaccinale_pct',
       'Tests_Realises', 'Distance_Centre_Sante_km', 'Reservoirs_Animaux',
       'Mobilite_Humaine', 'Cas_Confirmes', 'Saison', 'Population_Risque',
       'Taux_Positivite_pct']

for i in columns :
  dflm[i] = le.fit_transform(dflm[i])
```

```python
dflm
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
dflm_min = minmax.fit_transform(dflm)

dflm_scaler = pd.DataFrame(dflm_min, columns =dflm.columns)

dflm_scaler
```

```python
dflm_scaler.columns
```

```python
X = dflm_scaler.drop(
    ['Cas_Confirmes', 'ID'],
    axis=1
)

y = dflm_scaler['Cas_Confirmes']
```

```python
X
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=20)
```

```python
model_LRM = LinearRegression()
model_LRM.fit(x_train, y_train)
```

```python
bo = model_LRM.intercept_
bo
```

```python
coef = model_LRM.coef_
coef
```

```python
y_pred = model_LRM.predict(x_test)
y_pred
```

```python
print("MSE",mean_squared_error(y_test, y_pred))
print("MAE",mean_absolute_error(y_test, y_pred))
#print("RMSE",math.sqrt(mean_squared_error(y_test, y_pred)))
print("R2", r2_score(y_test, y_pred) )
```
