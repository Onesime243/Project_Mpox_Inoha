# Génère les prédictions du modèle sur les données de test (x_test) à l'aide du modèle de régression linéaire entraîné

print("MSE :", mean_squared_error(y_test, y_pred))

# Calcule et affiche l'Erreur Absolue Moyenne (MAE), qui indique l'écart moyen absolu entre les prédictions et les vraies valeurs

# Calcule et affiche la Racine de l'Erreur Quadratique Moyenne (RMSE) pour exprimer l'erreur dans la même unité que les données normalisées
print("RMSE :", math.sqrt(mean_squared_error(y_test, y_pred)))

# Calcule et affiche le Coefficient de Détermination (R²), qui mesure la proportion de la variance de la cible expliquée par le modèle

```python
print("MSE",mean_squared_error(y_test, y_pred))
print("MAE",mean_absolute_error(y_test, y_pred))
#print("RMSE",math.sqrt(mean_squared_error(y_test, y_pred)))
print("R2", r2_score(y_test, y_pred) )
```

# Interprétation des Résultats
1. Le Coefficient de Détermination ($R^2 = 0.9325$)Interprétation statistique : Le $R^2$ est proche de 1, ce qui indique une excellente performance globale. Très précisément, $93,25\%$ de la variance (la fluctuation) du nombre de cas confirmés est expliquée par les variables indépendantes incluses dans ton modèle de régression.Signification contextuelle : C'est un score particulièrement élevé pour des données réelles. Cela démontre que les variables explicatives sélectionnées (facteurs spatiaux, temporels, démographiques ou régionaux) capturent de manière très robuste la dynamique et l'intensité des cas.
2. L'Erreur Absolue Moyenne ($\text{MAE} = 0.0260$)Interprétation statistique : La MAE mesure la magnitude moyenne des erreurs de prédiction, sans tenir compte de leur direction.Signification contextuelle : Comme tes données ont été normalisées entre 0 et 1 avec le MinMaxScaler, une MAE de $0,026$ signifie qu'en moyenne, les prédictions du modèle ne s'écartent que de $2,6\%$ de la valeur réelle sur l'échelle standardisée. C'est le signe d'un modèle très précis qui colle de près à la réalité terrain.
3. L'Erreur Quadratique Moyenne ($\text{MSE} = 0.00157$)Interprétation statistique : La MSE mesure la moyenne des carrés des erreurs. Elle accorde un poids prépondérant aux grands écarts.Signification contextuelle : La valeur est extrêmement faible ($0,00157$). Cela t'indique que le modèle commet très peu de "grosses erreurs" ou d'aberrations majeures dans ses prédictions. Les résidus (les écarts entre les cas réels et les cas prédits) sont concentrés très près de zéro.
